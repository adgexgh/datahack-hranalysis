```python
df_new_feature=train.groupby('region').mean()[['previous_year_rating','avg_training_score','awards_won?','is_promoted']]
```

```python
empty_set_train_y=clf.predict(empty_set_train_x)
empty_set_test_y=clf.predict(empty_set_test_x)
```

```python
empty_set_train.loc[:,'predicted_previous_year_rating']=empty_set_train_y
empty_set_test.loc[:,'predicted_previous_year_rating']=empty_set_test_y
```

#### Now insert

```python
empty_set_train.drop(empty_set_train.columns.difference(['employee_id','predicted_previous_year_rating']), 1, inplace=True)
empty_set_test.drop(empty_set_test.columns.difference(['employee_id','predicted_previous_year_rating']), 1, inplace=True)
```

```python
train=train.merge(empty_set_train,'left', on='employee_id' )
test=test.merge(empty_set_test,'left', on='employee_id' )
```

```python
train['previous_year_rating'].fillna(train['predicted_previous_year_rating'],inplace=True)
test['previous_year_rating'].fillna(test['predicted_previous_year_rating'],inplace=True)
```

```python
train.head()
#drop 'employee_id','department','region','education','recruitment_channel','predicted_previous_year_rating'
```

### Split department, education,recruitment_channel into columns using pd.dummies

```python
train=pd.concat([train,pd.get_dummies(train.department,prefix='dep')],axis=1)
test=pd.concat([test,pd.get_dummies(test.department,prefix='dep')],axis=1)
```

```python
train=pd.concat([train,pd.get_dummies(train.education,prefix='educ')],axis=1)
test=pd.concat([test,pd.get_dummies(test.education,prefix='educ')],axis=1)
```

```python
train=pd.concat([train,pd.get_dummies(train.recruitment_channel,prefix='recr')],axis=1)
test=pd.concat([test,pd.get_dummies(test.recruitment_channel,prefix='recr')],axis=1)
```

### Drop excess columns

```python
train.drop(['department','region','education','recruitment_channel','predicted_previous_year_rating'],axis=1,inplace=True)
```

```python
test.drop(['department','region','education','recruitment_channel','predicted_previous_year_rating'],axis=1,inplace=True)
```

## Training model

```python
X_train_set=train.drop(['employee_id','is_promoted'],axis=1)
Y_train_set=train['is_promoted']
```
